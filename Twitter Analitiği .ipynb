## Bölüm Projesi : Takipçi Segmentasyonu 

Amacımız takipçileri segmente ederek hedef kullanıcılar belirlemek.Üretken olmak ve takipçi etkileşimi nedir ?

```python
df.head()
```

İlk olarak kullancı isimleri olan **screen_name** değişkeni **index** olarak geçireceğiz.

```python
df.index = df["screen_name"]
```

```python
df.head()
```

Segmentasyon datası için takipçi sayısını ve tweet sayısını alarak bir ayrıştırma yapıyoruz.

```python
seg_data = df[["followers_count","statuses_count"]]
```

```python
seg_data.info()
```

```python
seg_data
```

Bir ilkel segmentasyon işlemi yapacağız ve bu işlemi yaptığımızda neticesinde bir standartlaştırma oalcaktır.Fakat bazı takipçi değerlerinde görüldüğü üzere 0 değerlerine bulunmaktadır.Bu problemlenden kurtulabilmek adına yine de 0 sayısının etkisini de koruyabilmek adına ekleme işlemi gerçekleştireceğiz.

```python
seg_data["followers_count"] = seg_data["followers_count"] + 0.01
seg_data["statuses_count"] = seg_data["statuses_count"] + 0.01
```

```python
seg_data
```

Şimdi ki yapacağımız işlem bütün değerleri standartlaştırıp ondan sonra birbirleri ile çarptığımızda hedeflemiş olduğumuz tek bir skor değerine ulaşmış olsun.

```python
# Standartlaştırma İşlemi
seg_data = seg_data.apply(lambda x : (x-min(x)) / (max(x)- min(x)))
```

Standartlaştırma işleminden sonra da görüldüğü gibi bazı 0 değerleri gene bulunmaktadır.Şimdi yine üzerine ekleme işlemi gerçekleşleştirerek amacımıza hazır hale getireceğiz.

```python
seg_data["followers_count"] = seg_data["followers_count"] + 0.01
seg_data["statuses_count"] = seg_data["statuses_count"] + 0.01
```

```python
seg_data.head()
```

Şimdi bu değerleri birbirleri ile çarparak tek bir skor oluşturacağız.

```python
skor = seg_data["followers_count"] *  seg_data["statuses_count"]
```

```python
skor
```

Şimdi bu skorlara göre bu kullanıcıları sıralayıp belirlediğimiz bir kural ya da eşik değerine göre iki farklı sınıfa ayırma işlemi gerçekleştireceğiz.

```python
# Skor değerlerine göre sıralama
skor.sort_values(ascending = False)
```

```python
# Medyandan büyük olan değerleri alarak segmentasyon yapma -- İstenilen duruma göre skor içerisinde eşik belirlenebilir.
skor[skor > skor.median()].sort_values(ascending = False)
```

Segmente edilmiş datamız içerisine skor değişkenini yerleştiriyoruz.

```python
seg_data["skor"] = skor
```

```python
seg_data
```

Skor değişkenin yerleştirdikten sonra segmentleri bilgisini de aynı veri içerisine atarak belirlediğimiz eşik olan medyan değerine göre sınıflanmasını sağlayacağız.

```python
# Medyan değerine eşit ya da büyük olanlara A, bu değerden küçük olanlara ise B harfini tanımladık.
import numpy as np
seg_data["segment"] = np.where(seg_data["skor"] >= skor.median(), "A", "B")
```
