## B√∂l√ºm Projesi : Takip√ßi Segmentasyonu 

Amacƒ±mƒ±z takip√ßileri segmente ederek hedef kullanƒ±cƒ±lar belirlemek.√úretken olmak ve takip√ßi etkile≈üimi nedir ?

```python
df.head()
```

ƒ∞lk olarak kullancƒ± isimleri olan **screen_name** deƒüi≈ükeni **index** olarak ge√ßireceƒüiz.

```python
df.index = df["screen_name"]
```

Segmentasyon datasƒ± i√ßin takip√ßi sayƒ±sƒ±nƒ± ve tweet sayƒ±sƒ±nƒ± alarak bir ayrƒ±≈ütƒ±rma yapƒ±yoruz.

```python
seg_data = df[["followers_count","statuses_count"]]
```

```python
seg_data.info()
```

```python
seg_data
```

Bir ilkel segmentasyon i≈ülemi yapacaƒüƒ±z ve bu i≈ülemi yaptƒ±ƒüƒ±mƒ±zda neticesinde bir standartla≈ütƒ±rma oalcaktƒ±r.Fakat bazƒ± takip√ßi deƒüerlerinde g√∂r√ºld√ºƒü√º √ºzere 0 deƒüerlerine bulunmaktadƒ±r.Bu problemlenden kurtulabilmek adƒ±na yine de 0 sayƒ±sƒ±nƒ±n etkisini de koruyabilmek adƒ±na ekleme i≈ülemi ger√ßekle≈ütireceƒüiz.

```python
seg_data["followers_count"] = seg_data["followers_count"] + 0.01
seg_data["statuses_count"] = seg_data["statuses_count"] + 0.01
```

```python
seg_data
```

≈ûimdi ki yapacaƒüƒ±mƒ±z i≈ülem b√ºt√ºn deƒüerleri standartla≈ütƒ±rƒ±p ondan sonra birbirleri ile √ßarptƒ±ƒüƒ±mƒ±zda hedeflemi≈ü olduƒüumuz tek bir skor deƒüerine ula≈ümƒ±≈ü olsun.

```python
# Standartla≈ütƒ±rma ƒ∞≈ülemi
seg_data = seg_data.apply(lambda x : (x-min(x)) / (max(x)- min(x)))
```

Standartla≈ütƒ±rma i≈üleminden sonra da g√∂r√ºld√ºƒü√º gibi bazƒ± 0 deƒüerleri gene bulunmaktadƒ±r.≈ûimdi yine √ºzerine ekleme i≈ülemi ger√ßekle≈üle≈ütirerek amacƒ±mƒ±za hazƒ±r hale getireceƒüiz.

```python
seg_data["followers_count"] = seg_data["followers_count"] + 0.01
seg_data["statuses_count"] = seg_data["statuses_count"] + 0.01
```

```python
seg_data.head()
```

≈ûimdi bu deƒüerleri birbirleri ile √ßarparak tek bir skor olu≈üturacaƒüƒ±z.

```python
skor = seg_data["followers_count"] *  seg_data["statuses_count"]
```

```python
skor
```

≈ûimdi bu skorlara g√∂re bu kullanƒ±cƒ±larƒ± sƒ±ralayƒ±p belirlediƒüimiz bir kural ya da e≈üik deƒüerine g√∂re iki farklƒ± sƒ±nƒ±fa ayƒ±rma i≈ülemi ger√ßekle≈ütireceƒüiz.

```python
# Skor deƒüerlerine g√∂re sƒ±ralama
skor.sort_values(ascending = False)
```

```python
# Medyandan b√ºy√ºk olan deƒüerleri alarak segmentasyon yapma -- ƒ∞stenilen duruma g√∂re skor i√ßerisinde e≈üik belirlenebilir.
skor[skor > skor.median()].sort_values(ascending = False)
```

Segmente edilmi≈ü datamƒ±z i√ßerisine skor deƒüi≈ükenini yerle≈ütiriyoruz.

```python
seg_data["skor"] = skor
```

```python
seg_data
```

Skor deƒüi≈ükenin yerle≈ütirdikten sonra segmentleri bilgisini de aynƒ± veri i√ßerisine atarak belirlediƒüimiz e≈üik olan medyan deƒüerine g√∂re sƒ±nƒ±flanmasƒ±nƒ± saƒülayacaƒüƒ±z.

```python
# Medyan deƒüerine e≈üit ya da b√ºy√ºk olanlara A, bu deƒüerden k√º√ß√ºk olanlara ise B harfini tanƒ±mladƒ±k.
import numpy as np
seg_data["segment"] = np.where(seg_data["skor"] >= skor.median(), "A", "B")
```
